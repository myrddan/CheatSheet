---

### Objectif :

Utiliser les méthodes de la classe str pour nettoyer et formater un texte brut.

Consigne :

1. Créez une variable texte_brut contenant la chaîne suivante :

```python
"  python est un Langage de Programmation Puissant et facile à apprendre.  "
````

1. Effectuez les opérations suivantes pour mettre en forme ce texte :
    - Supprimez les espaces superflus au début et à la fin de la chaîne avec strip().
    - Convertissez tout le texte en minuscules avec lower().
    - Remplacez le mot "Puissant" par "polyvalent" avec replace().
    - Capitalisez la première lettre de chaque mot avec title().
    - Affichez le texte final après ces modifications.
<br>
1. Le texte final doit être stocké dans une variable nommée texte_formate.

Exemple de résultat attendu :

    Python Est Un Langage De Programmation Polyvalent Et Facile À Apprendre.


`````python
Utilisez strip() pour enlever les espaces en début et fin de chaîne.
Utilisez lower() pour convertir tout le texte en minuscules.
Utilisez replace() pour remplacer "Puissant" par "polyvalent".
Utilisez title() pour capitaliser chaque mot.
`````

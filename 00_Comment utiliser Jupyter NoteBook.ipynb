Jupyter Notebook est un environnement interactif open-source qui permet de créer et partager des documents combinant du code exécutable (Python, R, etc.), des visualisations, du texte formaté en Markdown et des équations, idéal pour l'analyse de données, la recherche et l'enseignement.

Le Markdown est un langage de balisage léger qui permet de formater du texte de manière simple et intuitive, idéal pour rédiger des documents structurés (comme des README, des notes ou des pages web) sans utiliser de code HTML complexe.

# Mise en place

Pour utiliser jupyter notebook, il faut soit utiliser Visual studio code  ou avoir la license pro de PyCharm.

Dans pycharm, installer la librairie "Jupyter" en suivant la procédure décrire dans le fichier 00_installation d'un package

# Méthode de fonctionnement

Dans Jupyter Notebook, les blocs (ou cellules) fonctionnent de manière indépendante : les cellules de code s’exécutent séquentiellement (ou dans l’ordre choisi) en conservant leur état en mémoire, tandis que les cellules Markdown servent à documenter et expliquer le code avec du texte formaté, le tout dans un même document interactif et modifiable à la volée.

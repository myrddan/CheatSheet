# Cheatsheet Markdown pour jupyter

Jupyter Notebook est un environnement interactif open-source qui permet de créer et partager des documents combinant du code exécutable (Python, R, etc.), des visualisations, du texte formaté en Markdown et des équations, idéal pour l'analyse de données, la recherche et l'enseignement.

Le Markdown est un langage de balisage léger qui permet de formater du texte de manière simple et intuitive, idéal pour rédiger des documents structurés (comme des README, des notes ou des pages web) sans utiliser de code HTML complexe.


# Titres (Headings)
   Le nombre de # en début de ligne va déterminer le niveau du titre :
- # Pour les titres
```markdown
# Pour les titres
```
- ## Pour les sous-titres
```markdown
## Pour les sous-titres
```
- ### pour les parties
```markdown
### pour les parties
```
- #### pour les sous-parties

```markdown
#### for 4th level subheadings
```

# Formatage (Emphasis)

Formatage du texte :

   ```__gras__ or **gras**```
- *Mise en italic* :
   ``` _italic_ or *italic*```
- Font monospace :
   ``` `monospace` ```

# Les listes

## A puce

Ajouter + ou - pour créer la liste, vous pouvez gérer les sous-niveau avec des espaces.

- Main bullet point
  -  Sub bullet point 2 spaces
    - Sub bullet point 4 spaces or tab
```
- Main bullet point
  -  Sub bullet point 2 spaces
    - Sub bullet point 4 spaces or tab
```

## Ordonné

Ajouter 1. ou 1) pour créer la liste, vous pouvez gérer les sous-niveaux avec des espaces.

1. Main point
    1) Sub point
    1) Sub point

```
1. Main point
    1) Sub point
    1) Sub point

```
